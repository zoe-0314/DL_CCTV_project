```python
# 풀어둔 동영상별 각 경로 구하기
# 동영상 폴더 구조에 따라 수정 필요

path_dir = '/content/ucf_file'
file_list = glob.glob(path_dir+'/*')
print(file_list)

# --- ucf-crime을 위한 코드 --

video_file_list = []
for each_file in file_list:
  if 'Anomaly' in each_file:
    each_folder = glob.glob(each_file+'/*')

    for each in each_folder:
      if 'Assault' in each:
        each = glob.glob(each+'/*')

        video_file_list = video_file_list + each
  else:
    normal_file = glob.glob(each_file+'/*')
    video_file_list = video_file_list + normal_file


print(video_file_list)
```

# 각 동영상별 npy 파일 생성

```python
from i3dFeatureExtraction import FeatureExtraction
from tqdm import tqdm
import torch
```
